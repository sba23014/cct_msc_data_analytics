####  R^2 Score
The amount of variance in the dependent variable—the one you're attempting to predict—that can be accounted for by the independent variables in the regression model is measured by the R-squared.

It is a number between 0 and 1, where 0 means that no variation in the dependent variable is explained by the independent variables, and 1 means that all of the variance is explained.

A higher R-squared value indicates that the model is a better fit for the data. R-squared assesses the model's goodness of fit.

####  Adjusted R^2 Score
Adjusted R-squared is a modification of R-squared that accounts for the number of independent variables in the model.

It penalises the insertion of extraneous variables whose performance is not appreciably enhanced by them. If more predictors are added to the model without significantly increasing its explanatory power, the adjusted R-squared value will decline.

When comparing models with varying numbers of predictors, adjusted R-squared provides a more trustworthy measure of the goodness of fit of the model.

It accounts for both the model's complexity and explained variance.

#### Making predictions using lm10

```python
# Making predictions using the model
X = media[['weekend','Character_A','ad_impression_million']]
X = sm.add_constant(X)
Predicted_views = lm_10.predict(X)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(media.Views_show, Predicted_views)
r_squared = r2_score(media.Views_show, Predicted_views)
```

```python
print('Mean_Squared_Error :' ,mse)
print('r_square_value :',r_squared)
```

```python
#Actual vs Predicted
c = [i for i in range(1,81,1)]
fig = plt.figure()
plt.plot(c,media.Views_show, color="blue", linewidth=2.5, linestyle="-")
plt.plot(c,Predicted_views, color="red",  linewidth=2.5, linestyle="-")
fig.suptitle('Actual and Predicted', fontsize=20)              # Plot heading 
plt.xlabel('Index', fontsize=18)                               # X-label
plt.ylabel('Views', fontsize=16)                               # Y-label
```

```python
# Error terms
c = [i for i in range(1,81,1)]
fig = plt.figure()
plt.plot(c,media.Views_show-Predicted_views, color="blue", linewidth=2.5, linestyle="-")
fig.suptitle('Error Terms', fontsize=20)              # Plot heading 
plt.xlabel('Index', fontsize=18)                      # X-label
plt.ylabel('Views_show-Predicted_views', fontsize=16)                # Y-label
```
