```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Fetch historical data for GSK plc
gsk = yf.Ticker("GSK")

# Get historical market data
hist = gsk.history(period="1y")  # Example: 1 year of data. Adjust as needed.

# Display the first few rows of the dataframe
print(hist.head())

# Plot the closing prices
plt.figure(figsize=(14, 7))
plt.plot(hist.index, hist['Close'], label='GSK Closing Price')
plt.title('GSK plc Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price (Currency)')
plt.legend()
plt.show()

# Calculate and plot a 30-day moving average of the closing prices
hist['30 Day MA'] = hist['Close'].rolling(window=30).mean()

plt.figure(figsize=(14, 7))
plt.plot(hist.index, hist['Close'], label='GSK Closing Price')
plt.plot(hist.index, hist['30 Day MA'], label='30 Day Moving Average', color='orange')
plt.title('GSK plc Closing Price and 30 Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Price (Currency)')
plt.legend()
plt.show()
```

```python
import yfinance as yf

# Define the ticker symbol for GSK plc
ticker_symbol = "GSK"

# Fetch the GSK data
gsk = yf.Ticker(ticker_symbol)

# Get historical market data, using 'max' for the maximum time period available
hist = gsk.history(period="max")

# Display the first few rows to confirm it's what you're looking for
print(hist.head())

# Save the dataframe to a CSV file if needed
hist.to_csv('GSK_stock_data.csv')
```
