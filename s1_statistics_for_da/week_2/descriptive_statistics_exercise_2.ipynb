## Import packages & load data

```python
# Import EDA & visualisation packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Configure default colour scheme for seaborn
sns.set(color_codes=True)

# Display all columns of the pandas df
pd.set_option('display.max_columns', None)

# Suppress warning messages
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read Excel file into Pandas DataFrame (df)
df = pd.read_excel('Exercise_2_Descriptive_Stats.xlsx')

# Return the first 5 rows
df.head()
```

```python
# Print the dimensionality of the df
df.shape
```

## Initial data exploration

```python
# Print a concise summary of the df
df.info()
```

```python
# List of columns to convert
cols_to_convert = ['Area_(ft.)', 'Price_in_euros']

# Convert selected columns to 'int'
df[cols_to_convert] = df[cols_to_convert].astype('int64')

# Validate the results
df.dtypes
```

```python
# Rename a single column
df.rename(columns = {'Year_of_sale': 'Sale_Year', 'Month_of_sale': 'Sale_Month', 'Type_of_property': 'Property_Type', 'Price_in_euros': 'Price_EUR'}, inplace=True)

# Validate changes
df.head()
```

```python
# Total number of rows and columns
print('df shape:')
print(df.shape)
print()

# Rows containing duplicate data
duplicate_rows_df = df[df.duplicated()]
print('Number of duplicate rows:')
print(duplicate_rows_df.shape)
print()

# Used to count the number of rows before removing the data
print('Row count before removing duplicates:')
print()
print(df.count())
print()

# Dropping the duplicates
df = df.drop_duplicates()

# Counting the number of rows after removing duplicates.
print('Row count after removing duplicates:')
print()
print(df.count())
print()
```

```python
# Finding the null values.
print(df.isnull().sum())
```

```python
# Dictionary to map integers to month names
month_dict = {
    1: 'Jan',
    2: 'Feb',
    3: 'Mar',
    4: 'Apr',
    5: 'May',
    6: 'Jun',
    7: 'Jul',
    8: 'Aug',
    9: 'Sep',
    10: 'Oct',
    11: 'Nov',
    12: 'Dec'
}

# Replace the integer values with month names
df['Sale_Month'] = df['Sale_Month'].replace(month_dict)

# Validate the results
df.head()
```

```python
# Filter rows where any value is null
null_rows_df = df[df.isnull().any(axis=1)]

# Review null rows to identify trends and potential reasons
null_rows_df
```

```python
# Generate descriptive statistics for numerical attributes
df_stats = pd.DataFrame(df.describe())

# Calculate Coefficient of Variation and add it as a new row
df_stats.loc['cv'] = (df_stats.loc['std'] / df_stats.loc['mean']) * 100

# Round the values to 3 decimal points
df_stats = df_stats.round(3)

# Validate the results
df_stats
```

**Count:** There are 267 rows, none of the numerical datatype columns appear to be missing values 

**Mean vs Median (50%):** the mean and median values are observed to be approximately equivalent excluding: 'Area_(ft.)' & 'Price_EUR'

```python
# Filter out the columns you don't want in the plot
filtered_df = df_stats.drop(['ID', 'Building', 'Property_#'], axis = 1)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(filtered_df.columns, filtered_df.loc['cv'])
plt.xlabel('Columns')
plt.ylabel('Coefficient of Variation (CV)')
plt.title('Coefficient of Variation (CV) for Selected Columns')
plt.show()
```
