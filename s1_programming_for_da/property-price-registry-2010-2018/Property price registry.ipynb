```python
# district 1 20 houses 0.5 mln -> 10 mln
```

```python
# 20 houses 0.5 mln -> 0.5 mln average
data_frame['COLUMN_NUMBER'] = [number for number in range(len(data_frame.index))]
```

```python
totals = data_frame.groupby("COUNTY").agg({"COLUMN_NUMBER": "count", "SALE_PRICE": "sum"})
totals = totals.reset_index().rename( columns = {"COLUMN_NUMBER": "Volume", "SALE_PRICE": "Total", "COUNTY": "County"})
```

```python

totals.sort_values(by=["Total"])
```
